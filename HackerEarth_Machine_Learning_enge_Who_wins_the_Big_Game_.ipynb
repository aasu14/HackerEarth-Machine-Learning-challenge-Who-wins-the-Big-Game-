```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import matplotlib as mpl

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
from sklearn.ensemble import ExtraTreesClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from lightgbm import LGBMClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold, learning_curve, ShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import VotingClassifier, IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import confusion_matrix
```

```
X = train.drop(columns=['ID', 'Won_Championship'])
y = train.Won_Championship
```

```
print('Percentage of each class in Target Variable \n')
print((train.Won_Championship.value_counts()/len(train))*100)
```

## Baseline Foundation

```
rfc = RandomForestClassifier(n_estimators=200, n_jobs=-1, random_state=10, max_features=0.3 , bootstrap=False)
cross_val_score(rfc, X, y, cv=2, n_jobs=-1, verbose=1, scoring='f1_weighted').mean()
```

## View Important Fetures

```
y = train.Won_Championship
corr = pd.concat(objs=[X, y], axis=1).corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```
X_test = test.drop(columns=['ID'])
concat = pd.concat(objs=[X, X_test], axis=0)
```

```
concat.drop(columns=['Previous_SB_Wins', 'Team_Value', 'Playing_Style'
                     ], inplace=True)

ss = StandardScaler()
concat = ss.fit_transform(concat)

X, X_test = concat[:len(train), :], concat[len(train):, :]
```

## Working

```
rfc = RandomForestClassifier(n_estimators=500, n_jobs=-1, random_state=30, max_features=0.3 , bootstrap=False)
cross_val_score(rfc, X, y, cv=4, n_jobs=-1, verbose=1, scoring='f1_weighted').mean()
```

```
nnc = MLPClassifier(hidden_layer_sizes=(10,5), max_iter=300,activation = 'relu',solver='adam', 
                           random_state=1, batch_size=50)
cross_val_score(nnc, X, y, cv=5, n_jobs=-1, verbose=1, scoring='f1_weighted').mean()
```

```
knc = KNeighborsClassifier(n_neighbors=21, n_jobs=-1)
cross_val_score(knc, X, y, cv=5, n_jobs=-1, verbose=1, scoring='f1_weighted').mean()
```

```
lgbmc = LGBMClassifier(random_state=2, n_estimators=100, colsample_bytree=0.9, 
                       max_depth=3, learning_rate=0.7, boosting_type='gbdt')
cross_val_score(lgbmc, X, y, cv=5, n_jobs=-1, verbose=1, scoring='f1_weighted').mean()
```

```
xgbc = XGBClassifier(seed=7, n_jobs=-1, n_estimators=100, random_state=0, max_depth=7, learning_rate=0.1)
cross_val_score(xgbc, X, y, cv=5, n_jobs=-1, verbose=1, scoring='f1_weighted').mean()
```

```
from sklearn.ensemble import VotingClassifier
ensemble = VotingClassifier(estimators=[('rfc', rfc), ('xgbc', xgbc), ('knc', knc),
                                        ('nnc', nnc), ('lgbmc', lgbmc)],
                                         voting='soft', n_jobs=-1)
cross_val_score(ensemble, X, y, cv=5, n_jobs=-1, verbose=1, scoring='f1_weighted').mean()
```

```
ensemble.fit(X, y)
j = ensemble.predict(X_test)
sub = pd.DataFrame(data={'ID' : test.ID, 'Won_Championship' : j})
sub.to_csv('C1.csv', index=False)
```
